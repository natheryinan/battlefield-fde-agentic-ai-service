```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from fde.factors.cross_section import CrossSectionFactors
```

```python
# 构造一个 demo 横截面数据（以后你会换成 Kaggle / 真实因子）
assets = ["CN_A1", "CN_A2", "CN_BANK", "US_SPY", "US_QQQ", "US_NVDA"]

scores = pd.DataFrame(
    {
        "value":    [0.10, 0.20, 0.15, 0.90, 1.20, 1.50],
        "momentum": [-0.10, 0.05, 0.00, 1.10, 1.30, 1.60],
    },
    index=assets,
)

scores
```

```python
factors = CrossSectionFactors(scores=scores)

# factor_name=None → 使用所有因子的均值
df_view = factors.cn_us_view(factor_name=None)
df_view
```

```python
print(df_view)
```

```python

y_map = {"CN": 0.0, "US": 1.0}
df_view["y"] = df_view["region"].map(y_map)

fig, ax = plt.subplots(figsize=(8, 4))

for region in ["CN", "US"]:
    sub = df_view[df_view["region"] == region]
    ax.scatter(sub["score"], sub["y"], label=region, alpha=0.7)

ax.set_yticks([0.0, 1.0])
ax.set_yticklabels(["CN", "US"])
ax.set_xlabel("Factor score (global scale)")
ax.set_title("Cross-sectional factor scores: CN vs US")
ax.legend()
ax.grid(True, axis="x", linestyle="--", alpha=0.3)

plt.tight_layout()
plt.show()
```
